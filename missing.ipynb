```python
import pandas as pd

titanic_df = pd.read_csv('train.csv')
```

```python
# Numpy에서 NaN : Null
titanic_df.isna().head(3) # 모든 열 값의 NaN 여부 확인
```

```python
titanic_df.isna().sum() # 각 열의 NaN 개수 확인인
```

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000') # 'Cabin' 열의 NaN을 'C000'으로 대체
titanic_df.head()
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum() 
```
